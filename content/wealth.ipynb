```python
import pandas as pd
import seaborn as sns
```

```python
wealth_df = pd.read_csv('data/lifetime_wealth.csv',index_col=0)
wealth_df.head()
```

```python
sns.relplot(data=wealth_df,x='year',y='wealth_lifetime',kind='line')
```

```python
sns.relplot(data=wealth_df,x='year',y='wealth_lifetime',
            col='type',
            kind='line')
```

```python
sns.relplot(data=wealth_df,x='year',y='wealth_lifetime',
            col='type',hue='race',
            kind='line')
```
